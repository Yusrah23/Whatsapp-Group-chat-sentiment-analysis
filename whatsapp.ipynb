```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read file line by line
f = open("whatsapp.txt", encoding="utf8")
data = f.readlines()
f.close()
```

```python
print('num lines: %s' %(len(data)))
```

```python
#remove first whatsapp info message
dataset = data[1:]
cleaned_data = []
for line in dataset:
    #split each column by delimeter
    date = line.split(",")[0]
    line2 = line[len(date):]
    time = line2.split("]")[0][2:]
    line3 = line2[len(time):]
    name =line3.split(":")[0][4:]
    line4 = line3[len(name):]
    message = line4[6:-1]
    
    print(date, time, name, message)
    cleaned_data.append([date,time,name,message])
```

```python
#create dataframe
df = pd.DataFrame(cleaned_data, columns= ['Date', 'Time', 'Name', 'Message'])
```

```python
#save to excel
df.to_excel('chat_history3.xlsx', index=False)
```

```python
#import from excel
df= pd.read_excel('chat_history_cleaned.xlsx')
```

```python
df.head()
```

```python
#import library for sentiment analysis
from textblob import TextBlob
```

```python
#getting polarity_score
polarity_score = []
for i in range(0,df.shape[0]):
    score = TextBlob(df.iloc[i][3])
    score1 = score.sentiment[0]
    polarity_score.append(score1)
    
#concatenating the score with thw dataframe
df = pd.concat([df,pd.Series(polarity_score)] , axis = 1)
```

```python
df.head()
```

```python
df.rename(columns={df.columns[4]: "Sentiment"}, inplace = True)
```

```python
#creating a new column to label the sentiment
filters = [
   (df.Sentiment > 0) ,
   (df.Sentiment < 0) ,
   (df.Sentiment == 0)
]
values = ["Positive", "Negative", "Neutral"]

df["status"] = np.select(filters, values)
df.tail()
```

```python
count = df['status'].value_counts()
#Using matplotlib donught chart
plt.pie(count, labels = count.index,autopct='%.0f%%',startangle = 90,
        counterclock = False);
plt.title('Message sentiment');
```
